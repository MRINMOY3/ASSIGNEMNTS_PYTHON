```python
print('---------------------')
n = 5
for i in range(5):
    for j in range(i+1):
        print("*",end=" ")
    print()
print('---------------------')


print('---------------------')
n = 5
for i in range(5):
    for j in range(i+1):
        print(i+1, end=" ")
    print()
print('---------------------')


print('---------------------')
n=5
for i in range(n):
    for j in range(i+1):
        print(j+1,end=" ")
    print()
print('---------------------')



print('---------------------')
n = 5
i = 0
num = i+2
for i in range(n):
    for j in range(i+1):
        print(num,end=" ")
        num=num+1
    print()
print('---------------------')


print('---------------------')
n = 5
for i in range(1,n+1):
    for j in range(i):
        if (i+j)%2 == 0:
            print(0,end=" ")
        else:
            print(1, end=" ")
    print()
print('---------------------')





print('---------------------')
n=5
i = 1
num = i
for i in range(n+1):
    for j in range(i):
        print(num, end = " ")
        num = num +1
    print()
print('---------------------')



print('---------------------')
n=5
num = n
for i in range(n):
    num = n
    for j in range(i+1):
        print(num,end = " ")
        num = num - 1
    print()
print('---------------------')

print('---------------------')
n = 5
num = n
for i in range(n):
    for j in range(i+1):
        print(num,end=" ")
    num = num-1
    print()
print('---------------------')



print('---------------------')
n = 5
for i in range(n+1):
    for j in range(n,i,-1):
        print("*",end=" ")
    print()
print('---------------------')




print('---------------------')
n = 5
for i in range(1,n):
    for j in range(n,i,-1):
        print(i,end = " ")
    print()
print('---------------------')



print('---------------------')
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print(i, end=" ")
    print()
print('---------------------')



print('---------------------')
i = 5
n = 1
while(i>=1):
    print(str(n)*i)
    if i <= 3:
        n = n-1
    else:
        n = n+1
    i = i-1
print('---------------------')



print('---------------------')
n = 5
for i in range(n):
    num = 1
    for j in range(i,n):
        print(num,end=" ")
        num = num+1
    print()
print('---------------------')




print('---------------------')
n = 5
num = n
for i in range(n):
    for j in range(i,n):
        print(num,end = " ")
    num = num-1
    print()
print('---------------------')



print('---------------------')
n = 5
for i in range(n):
    for j in range(n,i,-1):
        print(j,end=" ")
    print()
print('---------------------')



print('---------------------')
i = 1
while i<=5:
    s = "* "*i
    print(s.center(10))
    i = i+1
print('---------------------')
```
