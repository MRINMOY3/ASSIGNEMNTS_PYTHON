```python
"""
Bubble 
Insertion
Selection

Sorting Algorithms
"""
class BubbleSort:
    
    def __init__(self, size, arr):
        self.size = size
        self.arr = arr
        print("Welcome to Bubble Sort")
    
    def display(self):
        print(self.arr)
    
    def sorting_algorithm(self):
        for i in range(self.size - 1):
            for j in range(self.size - i - 1):
                if self.arr[j] > self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp


class InsertionSort:
    
    def __init__(self, size, arr):
        print("Welcome to insertion Sort")
        self.size = size
        self.arr = arr
    
    def display(self):
        print(self.arr)
    
    def sorting_algorithm(self):
        i:int
        j:int
        i=0
        j=0
        for i in range(self.size):
            j=i-1
            x=self.arr[i]
            while(j>-1 and self.arr[j]>x):
                self.arr[j+1]=self.arr[j]
                j-=1
            self.arr[j+1]=x

class SelectionSort:
    def __init__(self, size, arr):
        self.size = size
        self.arr = arr
    
    def display(self):
        print(self.arr)
    
    def sorting_algorithm(self):
        for i in range(self.size - 1):
            for j in range(self.size - i - 1):
                if self.arr[j] > self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp

print("The BUBBLE SORT ALGORITHM")
arr = []
a = int(input("Enter the size of the list: "))
for i in range(a):
    b = int(input(f"Enter the element ({i+1}) in the list: "))
    arr.append(b)




print("SORTING ALGORITHM")
arr = []
a = int(input("Enter the size of the list: "))
for i in range(a):
    b = int(input(f"Enter the element ({i+1}) in the list: "))
    arr.append(b)

b1 = BubbleSort(a, arr)
b1.display()
b1.sorting_algorithm()
b1.display()



print("SORTING ALGORITHM")
arr = []
a = int(input("Enter the size of the list: "))
for i in range(a):
    b = int(input(f"Enter the element ({i+1}) in the list: "))
    arr.append(b)
b2=InsertionSort(a,arr)
b2.display()
b2.sorting_algorithm()
b2.display()



print("SORTING ALGORITHM")
arr = []
a = int(input("Enter the size of the list: "))
for i in range(a):
    b = int(input(f"Enter the element ({i+1}) in the list: "))
    arr.append(b)
b3=SelectionSort(a,arr)
b3.display()
b3.sorting_algorithm()
b3.display()

```
