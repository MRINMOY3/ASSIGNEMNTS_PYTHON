## 4.	What is difference between ordered data type and unordered data type?


In Python, there are two types of data types: ordered and unordered. Ordered data types retain the order specified by the programmer, such as strings, lists, and tuples.1 On the other hand, unordered data types retain the order specified by the programmer, such as sets and dictionaries. Sets are an unordered collection of data type that is iterable, mutable, and has no duplicate elements.2 On the other hand, dictionaries are an unordered collection of data in a key:value pair form, enclosed in curly brackets. Tuple is an ordered collection of one or more data items, not necessarily of the same type, put in parentheses, while a dictionary Dict() object is an unordered collection of data in a key:value pair form.
